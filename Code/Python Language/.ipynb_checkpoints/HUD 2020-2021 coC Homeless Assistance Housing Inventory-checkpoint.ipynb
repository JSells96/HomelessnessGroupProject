```python
import pylab
import scipy.stats as stats
```

```python
coc2020_2021 = np.random.normal(loc = 20, scale=5, size=150)
stats.probplot(coc2020_2021, dist="norm", plot=pylab)
pylab.show()
```

```python
coc2020_2021 = np.random.normal(loc = 20, scale=5, size=150)
ax = sns.distplot(coc2020_2021)
```

```python
from scipy.stats import chisquare
import numpy as np
```

```python
coc2020_2021= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(coc2020_2021)
```

```python
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Not Normal Distribution')
```
