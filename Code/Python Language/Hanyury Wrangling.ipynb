Checking to see if the data has a normal distribution.

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
homeless_pop = np.random.normal(loc = 20, scale=5, size=150)
stats.probplot(homeless_pop, dist="norm", plot=pylab)
pylab.show()
```

```python
homeless_pop = np.random.normal(loc = 20, scale=5, size=150)
ax = sns.distplot(homeless_pop)
```

there is a normal distribution for the data of Homeless population in the United states. 

to test for normality using Chi-Square

```python
from scipy.stats import chisquare
import numpy as np
```

```python
homeless_pop= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(homeless_pop)
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Not Normal Distribution')
```
