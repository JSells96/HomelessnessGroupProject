### checking for Normal Distribution

```python
Homeless_Race = np.random.normal(loc = 20, scale=5, size=150)
stats.probplot(Homeless_Race, dist="norm", plot=pylab)
pylab.show()
```

```python
Homeless_Race = np.random.normal(loc = 20, scale=5, size=150)
ax = sns.distplot(Homeless_Race)
```

### Testing for Normality using Chi- Square

```python
Homeless_Race= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(Homeless_Race)
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Not Normal Distribution')
```
