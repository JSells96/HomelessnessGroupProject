How do the variables relate to each other?

```python
Income19Cor = pd.DataFrame(Income19Keeps, columns = ['TypeOfHousing19', 'RentOrOwn19', 'PersonsInHousehold19','HouseholdIncomeAmt19', 'Disability19', 'Unemployment19', 'Welfare19'])
Income19Cor
```

```python
Income19Cor.corr(method='pearson')
```

```python
sns.heatmap(Income19Cor.corr(), annot=True)
```

Looks like we've got high correlations between whether there is a claimant of disability in the home and unemployment claimants as well and those receiving welfare. These three variables do not seem to have a huge relation to others, but a mild correlation to the household income, which was assumed. The type pof housing and whether the tenants rent or own does not seem to have a huge relation to anything.

```python
Income20Cor = pd.DataFrame(Income20Keeps, columns = ['TypeOfHousing20', 'RentOrOwn20', 'PersonsInHousehold20','HouseholdIncomeAmt20', 'Disability20', 'Unemployment20', 'Welfare20'])
Income21Cor = pd.DataFrame(Income21Keeps, columns = ['TypeOfHousing21', 'RentOrOwn21', 'PersonsInHousehold21','HouseholdIncomeAmt21', 'Disability21', 'Unemployment21', 'Welfare21'])
```

```python
Income20Cor.corr(method='pearson')
```

```python
sns.heatmap(Income20Cor.corr(), annot=True)
```

These numbers look very similar to 2019. Let's check 2021 to see if we have a larger change.

```python
sns.heatmap(Income21Cor.corr(), annot=True)
```

No crazy shift through the years, looks as assumed.
