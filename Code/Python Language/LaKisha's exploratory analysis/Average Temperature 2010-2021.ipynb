# Checking for Normal Distribution

```python
import pylab
import scipy.stats as stats
```

```python
Temp = np.random.normal(loc = 20, scale=5, size=150)
stats.probplot(Temp, dist="norm", plot=pylab)
pylab.show()
```

```python
Temp = np.random.normal(loc = 20, scale=5, size=150)
ax = sns.distplot(Temp)
```

# Testing for normailty using Chi-Square

```python
from scipy.stats import chisquare
import numpy as np
```

```python
Temp= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(Temp)
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Not Normal Distribution')
```
