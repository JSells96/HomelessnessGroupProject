# Checking for Normal Distribution

```python
import pylab
import scipy.stats as stats
```

```python
Precipitation = np.random.normal(loc = 20, scale=5, size=150)
stats.probplot(Precipitation, dist="norm", plot=pylab)
pylab.show()
```

# Testing for Normality using Chi-Square

```python
from scipy.stats import chisquare
import numpy as np
```

```python
Precipitation= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(Precipitation)
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Not Normal Distribution')
```
