```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from scipy import stats
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data

```python
FU2010_2021 = pd.read_excel ('../../Data/2010-2021 Family Units.xlsx')
```

```python
FU2010_2021.head()
```

```python
FU2010_2021.info()
```

# Heat Map Correlation

```python
sns.heatmap(FU2010_2021.corr(), annot=True)
```

# Pearson Correlation

```python
FU2010_2021.corr(method = 'pearson').style.format("{:2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```

# Spearman Correlation

```python
FU2010_2021.corr(method = 'spearman').style.format("{:2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```

# The correlations shows from 2010-2013 more Family Units were available than 2017-2021.
