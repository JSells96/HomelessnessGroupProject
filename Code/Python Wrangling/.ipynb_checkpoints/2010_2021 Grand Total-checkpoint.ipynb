```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data

```python
GT2010_2021 = pd.read_excel ('../../Data/Grand Total HUD 2010-2021CoC Homeless Assistance Program Housing Inventory Report.xls')
```

```python
GT2010_2021.head
```

```python
GT2010_2021.info()
```

```python
GT2010_2021a= GT2010_2021.dropna()
GT2010_2021.head()
```

# Heat Map Correlation

```python
sns.heatmap(GT2010_2021a.corr(), annot=True)
```

# Pearson Correlation

```python
GT2010_2021a.corr(method='pearson')
```

```python
GT2010_2021a.corr(method = 'pearson').style.format("{:2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```

# Spearman Correlation

```python
GT2010_2021a.corr(method='spearman')
```

```python
GT2010_2021a.corr(method = 'spearman').style.format("{:2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```
