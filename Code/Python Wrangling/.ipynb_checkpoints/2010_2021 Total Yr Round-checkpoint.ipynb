```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data

```python
TYR2010_2021 = pd.read_excel ('../../Data/2010-2021 Total Year Round.xlsx')
```

```python
TYR2010_2021.head()
```

```python
TYR2010_2021.info()
```

# Heat Map Correlation

```python
sns.heatmap(TYR2010_2021.corr(), annot=True)
```

# Pearson Correlation

```python
TYR2010_2021.corr(method='pearson')
```

```python
TYR2010_2021.corr(method = 'pearson').style.format("{:2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```

# Spearman Correlation

```python
TYR2010_2021.corr(method='spearman')
```

```python
TYR2010_2021.corr(method = 'spearman').style.format("{:2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```
