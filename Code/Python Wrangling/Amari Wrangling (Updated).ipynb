```python
Trans2019_2020 = np.random.normal(loc = 20, scale=5, size=150)
stats.probplot(Trans2019_2020, dist="norm", plot=pylab)
pylab.show()
```

```python
Trans2019_2020= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(Trans2019_2020)
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Irregular Distribution')
```
