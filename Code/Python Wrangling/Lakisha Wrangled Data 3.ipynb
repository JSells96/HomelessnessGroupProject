```python
from scipy.stats import chisquare
import numpy as np
```

# There is a normal distribution for HUD 2017 - 2021 CoC Homeless Assistance Program Housing in the United states.
Test for normality using Chi-Square

```python
coc2017_2021= np.random.normal(loc = 20, scale=5, size=150)
statistic,pvalue=chisquare(coc2017_2021)
print('statistics=%.3f, p=%.3f\n' % (statistic, pvalue))
if pvalue > 0.05:
    print('Normal Distribution')
else:
    print('Not Normal Distribution')
```
